#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
min_prices = cinturon.groupby('level5')['price'].transform('min')
max_prices = cinturon.groupby('level5')['price'].transform('max')

# Aplicar la normalización Min-Max
cinturon['price_normalized'] = (cinturon['price'] - min_prices) / (max_prices - min_prices)

n_groups = cinturon['level5'].nunique()

fig, axes = plt.subplots(2, 2, figsize=(12, 10))  
axes = axes.flatten()  


for i, (group, data) in enumerate(cinturon.groupby('level5')):
    axes[i].hist(data['price_normalized'], bins=20, edgecolor='black', alpha=0.7)
    axes[i].set_title(f'Grupo: {group}')
    axes[i].set_xlabel('Precio Normalizado')
    axes[i].set_ylabel('Frecuencia')


plt.tight_layout()
plt.show()
```

getafe tiene los precios mas elevados a comparacion de los demas

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps

# Mapa centrado en (60 grados latitud y -2.2 grados longitud)
# Latitud, longitud
map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
cinturon.loc[:, "latitude"] = cinturon['latitude'].str.replace(',', '.', regex=False)
cinturon.loc[:, "longitude"] = cinturon['longitude'].str.replace(',', '.', regex=False)
```

```python
cinturon.loc[:, "latitude"] = cinturon['latitude'].astype(float)
cinturon.loc[:, "longitude"] = cinturon['longitude'].astype(float)
```

```python
coordenadas_dict = {}

# Agrupar por la columna 'level5'
for group, data in cinturon.groupby('level5'):
    coordenadas_dict[group] = list(zip(data['latitude'], data['longitude']))

print(coordenadas_dict)
```

```python
from ipyleaflet import Map, basemaps, Polyline

map = Map(center=(60, -2.2), zoom=2, min_zoom=1, max_zoom=20, basemap=basemaps.OpenStreetMap.Mapnik)

for key, coords in coordenadas_dict.items():
    polyline = Polyline(locations=coords, color="blue", weight=3)
    map.add_layer(polyline)

map
```
